# seqGANで使用するデータセットの作成

収集したレビューからseqGANでの学習に使用するデータセットの作成を行う

# 観光地のレビューから575の音韻となっている文字列を抽出する

```python
# ライブラリ

import MeCab
import glob
import numpy as np
import pickle
```

```python

tagger = MeCab.Tagger("-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd") #mecab neologd

places = ['hakataeki', 'uminaka', 'ohori', 'kaihinkouen', 'doom', 'dazaifu', 'all']

for place in places:
    with open('data/review/{}.txt'.format(place)) as f:
        s = f.read()

    text_list = s.split('。')

    haiku_list = []

    # print(text_list[0])


    for i in text_list:
        temp = i
        # print(temp)

        word_list = []

        # 形態素解析をして、単語毎にリストに格納
        for word in tagger.parse(temp).splitlines():
            word = word.replace('\t', ',')
            temp_word_list = word.split(',')
            word_list.append(temp_word_list)

        # 文末の記号を削除
        word_list.pop(-1)

        # print(word_list)

        # 格納された単語を繋げて17音の文字列を作る
        for j in range(len(word_list)):

            # 最初に空の文字列を用意
            temp_haiku = ""
            part = ""
            # 次の単語を示すカウント
            count = 0
            # 音を数えるcount
            haku_count = 0

            # 17音を超えるまで単語を繋げる

            if not ('名詞' ==  word_list[j+count][1] or '動詞' == word_list[j+count][1] or '形容詞' == word_list[j+count][1]):
                continue
            if ('名詞' ==  word_list[j+count][1] and ('非自立' == word_list[j+count][2] or '接尾' == word_list[j+count][2])):
                continue
            if ('動詞' ==  word_list[j+count][1] and '非自立' == word_list[j+count][2]):
                continue
            # 上の句5
            while haku_count < 5:
                # 単語を繋げる
                part =part + word_list[j + count][0]
                # 音を記録

                if word_list[j+count][-1] == "*":
                    break

                if '記号' not in word_list[j+count][1]:
                    temp = word_list[j+count][-1]
                    
                    # 小さい文字は音にならないので消去
                    # temp = temp.replace('ッ', '')
                    temp = temp.replace('ャ', '')
                    temp = temp.replace('ュ', '')
                    temp = temp.replace('ョ', '')
                    temp = temp.replace('ァ', '')
                    temp = temp.replace('ィ', '')
                    temp = temp.replace('ゥ', '')
                    temp = temp.replace('ェ', '')
                    temp = temp.replace('ォ', '')

                    haku_count += len(temp)
                # 元の文が終わっていたら終了
                if j + count >= len(word_list)-1:
                    break
                count += 1
                # print(part, haku_count)
            
            if haku_count == 5:
                temp_haiku += part
                # temp_haiku += "\n"
                part = ""
                haku_count = 0
            else:
                continue

            if not ('名詞' == word_list[j+count][1] or '動詞' == word_list[j+count][1] or '形容詞' == word_list[j+count][1]):
                continue
            if ('名詞' ==  word_list[j+count][1] and ('非自立' == word_list[j+count][2] or '接尾' == word_list[j+count][2])):
                continue
            if ('動詞' ==  word_list[j+count][1] and '非自立' == word_list[j+count][2]):
                continue


            # 中の句7
            while haku_count < 7:
                # 単語を繋げる
                part =part + word_list[j + count][0]
                # 音を記録
                
                if word_list[j+count][-1] == "*":
                    break
                if '記号' not in word_list[j+count][1]:
                    temp = word_list[j+count][-1]

                    # 小さい文字は音にならないので消去
                    # temp = temp.replace('ッ', '')
                    temp = temp.replace('ャ', '')
                    temp = temp.replace('ュ', '')
                    temp = temp.replace('ョ', '')
                    temp = temp.replace('ァ', '')
                    temp = temp.replace('ィ', '')
                    temp = temp.replace('ゥ', '')
                    temp = temp.replace('ェ', '')
                    temp = temp.replace('ォ', '')

                    haku_count += len(temp)
                # 元の文が終わっていたら終了
                if j + count >= len(word_list)-1:
                    break
                count += 1
                # print(temp_haiku, haku_count)
            
            if haku_count == 7:
                temp_haiku += part
                # temp_haiku += "\n"
                part = ""
                haku_count = 0
            else:
                continue

            if not ('名詞' == word_list[j+count][1] or '動詞' == word_list[j+count][1] or '形容詞' == word_list[j+count][1]):
                continue
            if ('名詞' ==  word_list[j+count][1] and ('非自立' == word_list[j+count][2] or '接尾' == word_list[j+count][2])):
                continue
            if ('動詞' ==  word_list[j+count][1] and '非自立' == word_list[j+count][2]):
                continue

            # 下の句5
            while haku_count < 5:
                # 単語を繋げる
                part =part + word_list[j + count][0]
                # 音を記録

                if word_list[j+count][-1] == "*":
                    break
                if '記号' not in word_list[j+count][1]:
                    temp = word_list[j+count][-1]

                    # 小さい文字は音にならないので消去
                    # temp = temp.replace('ッ', '')
                    temp = temp.replace('ャ', '')
                    temp = temp.replace('ュ', '')
                    temp = temp.replace('ョ', '')
                    temp = temp.replace('ァ', '')
                    temp = temp.replace('ィ', '')
                    temp = temp.replace('ゥ', '')
                    temp = temp.replace('ェ', '')
                    temp = temp.replace('ォ', '')

                    haku_count += len(temp)
                # 元の文が終わっていたら終了
                if j + count >= len(word_list)-1:
                    break
                count += 1
                # print(temp_haiku, haku_count)
            
            if haku_count == 5:
                temp_haiku += part
                # temp_haiku += "\n"
                temp_haiku += "\n"
                part = ""
                haku_count = 0
            else:
                continue
            

            if temp_haiku not in haiku_list:
                haiku_list.append(temp_haiku)

    # for i in haiku_list:
    #     print(i)
    print(len(haiku_list))

    temp = ""
    for i in haiku_list:
        temp += i
        # temp += "\n"

    haiku_output = temp


    # 出力
    with open('data/575/{}_575.txt'.format(place), mode='w') as f:
        f.write(haiku_output)
```

# レビューに出現した単語をidに変換する辞書を作成する

575の文字列を形態素解析して単語ごとにわかちがきを行う

```python

tagger = MeCab.Tagger('-Owakati -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd') #追加辞書を適用

places = ['hakataeki', 'uminaka', 'ohori', 'kaihinkouen', 'doom', 'dazaifu', 'all']

for place in places:

    with open('data/575/{}_575.txt' .format(place)) as f:
        s = f.readlines()

    # print(s[0])
    result = ""
    for line in s:

        # 前処理
        line = line.replace('、', '')
        line = line.replace('・', '')
        line = line.replace('.', '')

        # 分かち書き
        temp = tagger.parse(line)

        if  ")" in temp or "）" in temp or "（" in temp or "(" in temp or "「" in temp or "」" in temp :
            continue
        else:
            result += temp

    # print(result)

    with open('data/wakati/{}_575_wakati.txt'.format(place), mode='w') as f:
        f.write(result)
```

出現した単語を抽出し、それらに対応するidを与えて辞書を作成する

```python

temp_list = []

with open('data/wakati/all_575_wakati.txt') as f:
    s = f.readlines()


for line in s:
    line = line.split()
    temp_list.extend(line)

words = set(temp_list)
# print(words)

dict_count = 1
char_indices = {}  # 辞書初期化
indices_char = {}  # 逆引き辞書初期化

# 辞書の最初に文末記号を追加
char_indices = {"_" : 0}

for word in words:
    if not word in char_indices:  # 未登録なら
        char_indices[word] = dict_count  # 登録する      
        dict_count +=1
        # print(count,word)  # 登録した単語を表示
# 逆引き辞書を辞書から作成する
indices_char = dict([(value, key) for (key, value) in char_indices.items()])

hakataeki_dict = indices_char

# 作成した辞書はpickle形式で保存する
with open('data/dict/all_dict.pickle', 'wb') as f:
    pickle.dump(hakataeki_dict, f)

print(len(indices_char))
# print(indices_char)
```

# 作成した辞書に基づいて、575の文字列をid列に変換する

また、入力を１５単語の固定長になるようにパディングを行う

```python
# id列のデータをpaddingしないといけないみたいである
# 単語数が15に満たない場合は0を付け加える

places = ['hakataeki', 'uminaka', 'ohori', 'kaihinkouen', 'doom', 'dazaifu', 'all']

for place in places:

    # 作成した辞書の読み込み
    with open('data/dict/all_dict.pickle'.format(place), 'rb') as f:
        indices_char = pickle.load(f)
    char_indices = dict([(value, key) for (key, value) in indices_char.items()])

    # わかちがきした575の文字列の読み込み
    with open('data/wakati/{}_575_wakati.txt'.format(place)) as f:
        s = f.readlines()

    # １５単語になるようにパディング
    result = ""
    eos = 0
    for line in s:
        line = line.split()
        temp_length = len(line)
        line = np.pad(line, [0,15 - temp_length], 'constant', constant_values=(eos))

        # 単語に対応するidを並べて行く
        print(line)
        for i in line:
            # print(char_indices[i])
            if i == '0':
                result += str(char_indices["_"])
            else:
                # print(i)
                result += str(char_indices[i])
            result += " "

        result += "\n"

    print(result)

    with open('data/id/{}_575_id.txt'.format(place), mode='w') as f:
        f.write(result)
```
